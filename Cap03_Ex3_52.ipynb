Um turista está fugindo de uma onça no Pantanal, correndo em linha reta até seu carro com uma velocidade (constante) de 4.0 m.s$^{-1}$. Seu carro está a uma distância $d$. A onça está 26 m atrás do turista e correndo a 6.0 m.s$^{-1}$. O turista consegue alcançar seu carro em segurança. Qual o maior valor possível para $d$?

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
#x(t) do MRU
def xx(tt, x0, vx0):
    return x0 + vx0 * tt
```

```python
ttarr = np.linspace(0.0, 15.0, 100)
xxvec = np.vectorize(xx)
ttarr
```

```python
#MRU do turista
xxtur = xxvec(ttarr, x0=0.00, vx0=4.00)
xxtur
```

```python
#MRU da onça
xxonc = xxvec(ttarr, x0=-26.0, vx0=6.00)
xxonc
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.gca()

ax.plot(ttarr, xxtur, color='red', linestyle='-', label='turista')
ax.plot(ttarr, xxonc, color='green', linestyle='-', label='onça')
ax.axhline(y=52.0, color='black', linestyle ="--", label='carro')

#limites, formatação dos ticks e label dos eixos
ax.set_xlim(0.0, 16.0)
ax.xaxis.set_major_locator(mtick.MultipleLocator(2.0))
ax.xaxis.set_major_formatter(mtick.FormatStrFormatter('%.2f'))
ax.xaxis.set_minor_locator(mtick.MultipleLocator(1.0))
ax.set_xlabel('$t$ (s)')
ax.set_ylim(-26.0, 60.0)
ax.yaxis.set_major_locator(mtick.MultipleLocator(10.0))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.2f'))
ax.yaxis.set_minor_locator(mtick.MultipleLocator(5.0))
ax.set_ylabel('$x$ (m)')

ax.legend()
fig.tight_layout()
plt.show()
```
