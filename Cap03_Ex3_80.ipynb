Um gato está dormindo no chão, no meio de uma sala de 3.0 m de comprimento, quando um cachorro raivoso entra com uma velocidade de 1.50 m.s$^{-1}$. Imediatamente o gato acelera a 0.85 m.s$^{-2}$ em direção a uma janela aberta no lado oposto da sala. O cachorro, ficando receoso com a possibilidade real de alcançar o gato, começa a desacelerar a 0.10 m.s$^{-2}$ assim que entra na sala. O cachorro conseguirá alcançar o gato antes que ele pule pela janela?

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
#x(t) do MRUV
def xx(tt, x0, vx0, aa):
    return x0 + vx0 * tt + 0.5 * aa * tt ** 2
```

```python
ttarr = np.linspace(0.0, 2.5, 100)
xxvec = np.vectorize(xx)
ttarr
```

```python
#MRUV do cachorro
xxdog = xxvec(ttarr, x0=0.00, vx0=1.50, aa=-0.10)
xxdog
```

```python
#MRUV do gato
xxcat = xxvec(ttarr, x0=1.50, vx0=0.00, aa=0.85)
xxcat
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.gca()

ax.plot(ttarr, xxdog, color='red', linestyle='-', label='cachorro')
ax.plot(ttarr, xxcat, color='green', linestyle='-', label='gato')
ax.axhline(y=3.0, color='black', linestyle ="--", label='janela')

#limites, formatação dos ticks e label dos eixos
ax.set_xlim(0.0, 2.5)
ax.xaxis.set_major_locator(mtick.MultipleLocator(0.2))
ax.xaxis.set_major_formatter(mtick.FormatStrFormatter('%.2f'))
ax.xaxis.set_minor_locator(mtick.MultipleLocator(0.1))
ax.set_xlabel('$t$ (s)')
ax.set_ylim(0.0, 3.2)
ax.yaxis.set_major_locator(mtick.MultipleLocator(0.5))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.2f'))
ax.yaxis.set_minor_locator(mtick.MultipleLocator(0.25))
ax.set_ylabel('$x$ (m)')

ax.legend()
fig.tight_layout()
plt.show()
```
